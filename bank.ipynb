```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
bank=pd.read_excel('bank-full.xlsx')
bank
```

```python
#EDA
bank.info()
```

```python
# One-Hot Encoding of categrical variables
data1=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome','month'])
data1
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
data1
```

```python
data1.info()
```

```python
#binary encoding of binary output variables
# Custom Binary Encoding of Binary o/p variables 
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

```python
#model building
# Dividing our data into input and output variables
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.iloc[:,10]
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
#Model Predictions
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_labels':y_pred})
y_pred_df
```

```python
## Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
classifier.score(x,y)
```

```python
classifier.predict_proba(x)[:,1] 
#[:,1] -->is the threshold value
#if accuracy is is less than 0.5-->threshold value-->[:,0]
#if accuracy is is greater than 0.5-->threshold value-->[:,1]
#in this case the accuracy is 0.89 ...its greater than 0.5...sothreshold value-->[:,1]
```

```python
y_df=pd.DataFrame({'actaul_y':y,'y_pred_probabilities':classifier.predict_proba(x)[:,1] ,'y_pred_labels':y_pred})
y_df
```

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='blue')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='blue',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
